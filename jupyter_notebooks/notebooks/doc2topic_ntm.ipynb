```python
from doc2topic import models, corpora
import sys
from os import path

models.init_tf_memory()

data = corpora.DocData(path.expanduser('~/work/data/data.txt'))
```

```python
data.idx2token
```

```python
model = models.Doc2Topic()
model.build(data, n_topics=35, batch_size=1024*6, n_epochs=20, lr=0.05, l1_doc=0.000002, l1_word=0.000000015, word_dim=50, generator=None)

model.print_topic_words()
```

```python
model.history.history['fmeasure'][-1]
```

```python
model.save('ntm')
```

```python
topics = []
for topic in list(model.get_topic_words().values()):
    words = []
    for word in topic:
        words.append(word[0])
    topics.append(words)
```

```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.hashdictionary import HashDictionary

dictionary = HashDictionary(data.docs)
coherence_model = CoherenceModel(topics = topics, texts = data.docs, dictionary = dictionary, coherence='c_v')
coherence_per_topic = coherence_model.get_coherence_per_topic()
```
